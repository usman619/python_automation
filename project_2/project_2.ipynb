# Web Automation & Web Scraping

## XPath
Basic syntax:
``` xpath
//tagName
//tagName[1]
//tagName[@AttributeName="Value"]
```

## XPath Functions
- contains()
- starts-with()
``` xpath
//tagName[contains(@AttributeName,"Value")]
```

## XPath Opertors
- and
- or
``` xpath
//tagName[(expression1) and (expression2)]
//tagName[(expression1) or (expression2)]
```

XPath Playground Link: [https://scrapinghub.github.io/xpath-playground/](https://scrapinghub.github.io/xpath-playground/)

Example HTML:
Input:
``` html
<article class="main-article">
<h1> Titanic(1997) </h1>
<p class="plot" > 84 years later... </p>
<p class="plot2" > In the end ... </p>
<div class="full-script">
"13 meters. You should see it. "
"Okay, take her up and over the bow rail. "
</div>
</article>
```
Filters:<br>
XPath:
``` xpath
\\div\text()
```
Result:
``` text
"13 meters. You should see it. " "Okay, take her up and over the bow rail. " 
```
---
XPath:
``` xpath
\\p
```
Result:
``` text
 84 years later...
In the end ... 
```
---
XPath:
``` xpath
//p[1]
```
Result:
``` text
<p class="plot"> 84 years later... </p>
```
---

XPath:
``` xpath
//p[@class="plot"]/text()
```
Result:
``` text
84 years later...
```
---

XPath:
``` xpath
//div[@class="full-script"]/text()
```
Result:
``` text
 "13 meters. You should see it. " "Okay, take her up and over the bow rail. " 
```
---
XPath:
``` xpath
//p[(@class="plot")or(@class="plot2")]/text()
```
Result:
``` text
84 years later...
In the end ... 
```
---
XPath:
``` xpath
//div[contains(@class,"full-script")]/text()
```
Result:
``` text
"13 meters. You should see it. " "Okay, take her up and over the bow rail. " 
```

### Special Characters:
- /: Select the children from the node set on the left side of this character
- //: Specifies that the matching node set should be located at any level within the document
- .: Specifies the current context should be used (refers to present node)
- .. : Refers to a parent node
- *: Select all the children nodes considering the current context
- @: Select an attribute
- (): Grouping an XPath expression
- [n]: Indicates that a node with index "n" should be selected
XPath:
``` xpath
//article/h1/text()
```
Result:
``` text
Titanic(1997) 
```
---
XPath:
``` xpath
//article//text()
```
Result:
``` text

Titanic(1997)
84 years later...
In the end ...
"13 meters. You should see it. " "Okay, take her up and over the bow rail. " 
```
---
XPath:
``` xpath
//h1/.
```
Result:
``` html
<h1> Titanic(1997) </h1>
```
---
XPath:
``` xpath
//h1/..
```
Result:
``` html
<article class="main-article"> <h1> Titanic(1997) </h1> <p class="plot"> 84 years later... </p> <p class="plot2"> In the end ... </p> <div class="full-script"> "13 meters. You should see it. " "Okay, take her up and over the bow rail. " </div> </article> 
```
